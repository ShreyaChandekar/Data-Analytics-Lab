# Load data from Files

### Import Pandas Library

```python
import pandas as pd
```

### Load the data from CSV File

```python
# Read a CSV file
df_csv = pd.read_csv('House_Data.csv')
```

```python
# First 5 Rows of Data Frame
df_csv.head()
```

```python
# Struture of the Data frame
df_csv.info()
```

```python
# Read a CSV file using header paramenters
df_csv = pd.read_csv('House_Data.csv', header=0)
df_csv.head()
```

```python
# Read a CSV file using header and usecols paramenters
df_csv = pd.read_csv('House_Data.csv', header=0, usecols=["id", "price", "bedrooms"])
df_csv.head()
```

```python
# Read a CSV file using header, usecols and index_col paramenters
df_csv = pd.read_csv('House_Data.csv', header=0, usecols=["id", "price", "bedrooms"], index_col=['id'])
df_csv.head()
```

```python
# Read a CSV file using header, usecols, index_col and nrows paramenters
df_csv = pd.read_csv('House_Data.csv', header=0, usecols=["id", "price", "bedrooms"], index_col=['id'], nrows=10)
df_csv.head()
```

```python
# Read a CSV file using header, usecols, index_col, nrows and skiprows paramenters
df_csv = pd.read_csv('House_Data.csv', header=0, usecols=["id", "price", "bedrooms"], index_col=['id'], nrows=10, skiprows=[2,4])
df_csv.head()
```
